```python
miss = (df.isnull().sum().reset_index()
         .rename(columns={"index":"column",0:"missing"}))
miss["pct"] = (miss["missing"]/len(df)*100).round(2)
miss = miss[miss["missing"]>0].sort_values("pct",ascending=False)
print(miss.to_string(index=False))
```

```python
HIGH_MISS = 60.0
high_miss_cols = miss[miss["pct"]>=HIGH_MISS]["column"].tolist()
print("Dropping:", high_miss_cols)
df.drop(columns=[c for c in high_miss_cols if c in df.columns], inplace=True)

df[["killed","wounded"]] = df[["killed","wounded"]].fillna(0) if all(c in df.columns for c in ["killed","wounded"]) else df.get(["killed","wounded"],pd.DataFrame()).fillna(0)

for coord in ["latitude","longitude"]:
    if coord in df.columns:
        df[coord] = df.groupby("country")[coord].transform(lambda x: x.fillna(x.median()))
        df[coord].fillna(df[coord].median(), inplace=True)

cat_cols = df.select_dtypes("object").columns.tolist()
df[cat_cols] = df[cat_cols].fillna("Unknown")
if "day" in df.columns: df["day"].replace(0,1,inplace=True)

print(f"Missing cells after cleaning: {df.isnull().sum().sum()}")
```

Feature engineering is the process of transforming raw data into meaningful features that improve the performance of machine learning models. It involves techniques such as handling missing values, encoding categorical variables, scaling numerical features, and creating new variables from existing data.

```python
df['killed'] = df['killed'].fillna(0)
df['wounded'] = df['wounded'].fillna(0)


df['total_casualties'] = df['killed'] + df['wounded']

print(df[['killed', 'wounded', 'total_casualties']].head())
```

```python
def classify_severity(x):
    if x == 0:
        return 'No Casualty'
    elif x <= 5:
        return 'Low'
    elif x <= 20:
        return 'Medium'
    else:
        return 'High'

df['severity'] = df['total_casualties'].apply(classify_severity)

print(df[['total_casualties', 'severity']].head())
```

Outlier Detection & Capping using the 3×IQR method is a technique to handle extreme values in a dataset. The Interquartile Range (IQR) is calculated as the difference between the third quartile (Q3) and the first quartile (Q1). Any data points lying below Q1 − 3×IQR or above Q3 + 3×IQR are considered extreme outliers. Instead of removing them, capping replaces these values with the respective lower or upper bounds(winsorization).

```python

for col in ["killed","wounded","casualties"]:
    if col not in df.columns: continue
    Q1,Q3 = df[col].quantile(0.25), df[col].quantile(0.75)
    IQR   = Q3-Q1; upper = Q3+3*IQR
    print(f"{col}: IQR={IQR:.1f}, cap={upper:.0f}, outliers={(df[col]>upper).sum()}")
    df[f"{col}_capped"] = df[col].clip(upper=upper)
```

Now,we perform Hypothesis Testing

We test the claim:

H0 (Null Hypothesis): Average number of people killed per attack = 2
H1 (Alternative Hypothesis): Average number of people killed per attack ≠ 2

We will use:
✔ One-Sample T-Test (because population standard deviation is unknown)

Key Concepts:
- p-value: Probability of observing the data assuming H0 is true
- alpha (α): Significance level (we use 0.05)
- Decision:
    If p-value ≤ α → Reject H0
    If p-value > α → Fail to Reject H0

```python
import scipy.stats as stats

df = df[['killed']]

df = df.dropna()

df = df[df['killed'] >= 0]

print("Cleaned Data Summary:")
print(df.describe())

plt.figure(figsize=(8,5))
sns.histplot(df['killed'], bins=50, kde=True)
plt.title("Distribution of People Killed per Attack")
plt.xlabel("Number of People Killed")
plt.ylabel("Frequency")
plt.show()

sample = df['killed']

mu = 2

t_stat, p_value = stats.ttest_1samp(sample, mu)

print("\nT-statistic:", t_stat)
print("p-value:", p_value)

alpha = 0.05

if p_value <= alpha:
    print("\nDecision: Reject Null Hypothesis (H0)")
else:
    print("\nDecision: Fail to Reject Null Hypothesis (H0)")


x = np.linspace(-5, 5, 1000)
y = stats.t.pdf(x, df=len(sample)-1)

plt.figure(figsize=(10,6))
plt.plot(x, y)

plt.axvline(t_stat, linestyle='--', label='t-statistic')

plt.title("T-Distribution with T-Statistic")
plt.xlabel("t values")
plt.ylabel("Density")
plt.legend()
plt.show()

confidence = 0.95
mean = np.mean(sample)
sem = stats.sem(sample)

interval = stats.t.interval(confidence, len(sample)-1, loc=mean, scale=sem)

print("\nSample Mean:", mean)
print("95% Confidence Interval:", interval)
```

"Chi-square test helps identify whether categorical variables such as attack type and severity are dependent on each other.”

```python
from scipy.stats import chi2_contingency

table = pd.crosstab(geo_df['attack_type'], geo_df['severity'])

chi2, p, dof, expected = chi2_contingency(table)

print("Chi-square:", chi2)
print("p-value:", p)

if p <= 0.05:
    print("Variables are dependent")
else:
    print("Variables are independent")
```

“The Pearson correlation test confirms whether the relationship between casualties and injuries is statistically significant, supporting the observed trends in scatter plots."

```python
from scipy.stats import pearsonr

x = geo_df['killed']
y = geo_df['wounded']

data = pd.concat([x, y], axis=1).dropna()

corr, p_value = pearsonr(data['killed'], data['wounded'])

print("Correlation coefficient:", corr)
print("p-value:", p_value)


alpha = 0.05
if p_value <= alpha:
    print("Statistically significant correlation")
else:
    print("No significant correlation")
```

Mathematical equation:y=β0​+β1​x1​+β2​x2​+⋯+βn​xn​

```python
import numpy as np
import matplotlib.pyplot as plt


df_vis = geo_df[['wounded', 'killed']].dropna()

X = df_vis['wounded'].values
y = df_vis['killed'].values

X_mat = np.c_[np.ones(len(X)), X]


beta = np.linalg.inv(X_mat.T @ X_mat) @ X_mat.T @ y

y_pred = X_mat @ beta

plt.figure(figsize=(8,5))


plt.scatter(X, y, alpha=0.5, label='Data Points')

sorted_idx = np.argsort(X)
plt.plot(X[sorted_idx], y_pred[sorted_idx], linewidth=2, label='Regression Line')

plt.title("Least Squares Regression Line")
plt.xlabel("Number of Wounded")
plt.ylabel("Number of Killed")
plt.legend()
plt.show()
```

.MSE (Mean Squared Error) → lower is better

```python
from sklearn.metrics import mean_squared_error, r2_score

print("MSE:", mean_squared_error(y, y_pred))
print("R2 Score:", r2_score(y, y_pred))
```

“Ridge regression reduces overfitting by shrinking coefficients, while Lasso regression performs feature selection by forcing some coefficients to zero.”
->Ridge Regression (L2 Regularization):
Adds penalty on large coefficients
Prevents overfitting
Loss=RSS+λ∑β^2
->Lasso Regression (L1 Regularization):
Shrinks some coefficients to zero
Performs feature selection
Loss=RSS+λ∑∣β∣


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score

df_vis = geo_df[['wounded', 'killed']].dropna()

X = df_vis[['wounded']].values
y = df_vis['killed'].values


ridge = Ridge(alpha=1.0)
lasso = Lasso(alpha=0.1)

ridge.fit(X, y)
lasso.fit(X, y)


y_pred_ridge = ridge.predict(X)
y_pred_lasso = lasso.predict(X)

print("Ridge MSE:", mean_squared_error(y, y_pred_ridge))
print("Ridge R2:", r2_score(y, y_pred_ridge))

print("Lasso MSE:", mean_squared_error(y, y_pred_lasso))
print("Lasso R2:", r2_score(y, y_pred_lasso))

plt.figure(figsize=(10,6))

plt.scatter(X, y, alpha=0.4, label='Data Points')


sorted_idx = np.argsort(X[:,0])
X_sorted = X[sorted_idx]


plt.plot(X_sorted, ridge.predict(X_sorted), linewidth=2, label='Ridge Line')

plt.plot(X_sorted, lasso.predict(X_sorted), linewidth=2, linestyle='--', label='Lasso Line')

plt.title("Ridge vs Lasso Regression Line")
plt.xlabel("Number of Wounded")
plt.ylabel("Number of Killed")
plt.legend()
plt.show()
```

## Key Conclusions from the Analysis

### 📈 1. Terrorism Has Grown Dramatically
Terrorist attacks were relatively rare in the 1970s and 1980s.
A sharp and sustained surge began after **2010**, with **2014
recording the highest number of attacks** in the entire dataset.
This growth was primarily driven by the rise of **ISIS in the
Middle East** and the **Taliban in South Asia**.

### 🌍 2. Geography is Highly Concentrated
Terrorism is not a globally distributed phenomenon — it is
**heavily concentrated**. The **Middle East, South Asia, and
Sub-Saharan Africa** account for over **60% of all incidents**.
**Iraq, Afghanistan, and Pakistan** consistently rank as the
three most affected countries across all metrics.

### 💣 3. Bombing is the Weapon of Choice
**Bombing and Explosion** is by far the most frequently used
attack method, accounting for nearly **50% of all incidents**.
It is also responsible for the highest number of casualties,
making it the single greatest tactical threat in the dataset.

### 👥 4. A Few Groups Cause Most of the Damage
Despite thousands of groups being recorded, a small number
dominate the data. **ISIS (ISIL), Taliban, and Shining Path**
are consistently among the most active and most lethal groups —
confirming the **80/20 principle** in terrorism: a minority of
actors cause the majority of harm.

### 📅 5. Temporal Patterns Exist
Monthly analysis reveals **slightly elevated activity between
July and September**, suggesting seasonal patterns that could
inform preemptive security measures. Suicide attacks peaked
sharply around **2014–2015** before declining.

### 🏥 6. Human Cost is Staggering
The dataset records hundreds of thousands of deaths and injuries.
**Mass-casualty events** (25+ casualties) cluster heavily in the
**post-2000 era**, reflecting increasingly indiscriminate tactics
targeting civilian populations.
