```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('apple.csv')
df
```

```
df.info()
```

```
df.head(10)
```

```
#question number 1
top_rating= df.nlargest(5,'Star Rating')[['Product Name','Star Rating']]
print(top_rating)
```

```
#question number 2
lowest_rating= df.nsmallest(5,'Star Rating')[['Product Name','Star Rating']]
print(lowest_rating)
```

```
#question 3 :Ratings provided to top 5 star rating apple phones
top_rating= df.nlargest(5,'Star Rating')[['Number Of Ratings', 'Star Rating']]
print(top_rating)
```

```
#question 4 :Ratings provided to lowest 5 star rating apple phones
lowest_rating= df.nsmallest(5,'Star Rating')[['Number Of Ratings', 'Star Rating']]
print(lowest_rating)
```

```
#question 5 :Apple phone have highest number of reviews
highest_review= df.nlargest(1,'Number Of Reviews')[['Product Name','Number Of Reviews']]
print(highest_review)
```

```
highest_review = df.loc[df['Number Of Reviews'].idxmax(),['Product Name']]
print(highest_review)
```

```
#question 6 :Apple phone have lowest number of reviews
lowest_review= df.nsmallest(1,'Number Of Reviews')[['Product Name','Number Of Reviews']]
print(lowest_review)
```

```
y=df['Number Of Reviews'].sort_values().head(10)
x=df['Product Name'].head(10)
plt.barh(x,y)
plt.show()
```

```
#question 7 :Relationship between sale price and number of ratings
y1=df['Sale Price']
y2=df['Number Of Ratings']
x=df['Product Name']
colors = ['red', 'green']
plt.bar(x,y1,color=colors)
plt.bar(x,y2,color=colors)
plt.show()
```

```
y = df['Number Of Ratings'].head()
mylables = df['Product Name'].head()
y = y.dropna()
plt.pie(y, labels=mylables)
plt.show()
```

```
selected_product_names = df['Product Name'].isin(['APPLE iPhone 11 Pro Max (Gold, 256 GB)', 'APPLE iPhone 12 Pro (Silver, 512 GB)'])
filtered_x = df[selected_product_names]

plt.figure(figsize=(10, 6))
bar_width = 0.30
index = np.arange(len(filtered_x))
plt.bar(index-bar_width/2, filtered_x['Sale Price'], bar_width, label='Sale Price', color='blue')
plt.bar(index + bar_width/2, filtered_x['Mrp'], bar_width, label='Mrp',color='red')
plt.title('sales price vs number of ratings for Selected product name')
plt.xlabel('Product names')
plt.ylabel('divisons')
plt.xticks(index, filtered_x['Product Name'])
plt.legend()

plt.show()
plt.figure(figsize=(10, 6))
bar_width = 0.30
index = np.arange(len(filtered_x))
plt.bar(index-bar_width/2, filtered_x['Sale Price'], bar_width, label='Sale Price', color='blue')
plt.bar(index + bar_width/2, filtered_x['Mrp'], bar_width, label='Mrp',color='red')
plt.title('sales price vs number of ratings for Selected product name')
plt.xlabel('Product names')
plt.ylabel('divisons')
plt.xticks(index, filtered_x['Product Name'])
plt.legend()

plt.show()
```
